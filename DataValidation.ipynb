```python
df.plot(x="RPM", y="SPEED", kind="scatter", title='RPM Versus Speed Scatter Plot')
# plt.show()
# remove plot-file-name.png before trying to save.
# plt.savefig('scatter-RPM-SPEED.png')
plt.savefig('scatter-RPM-SPEED.png', 
                          dpi=300,
                          facecolor='w',
                          edgecolor='w',
                          orientation='landscape',
                          format=None,
                          transparent=True,
                          bbox_inches=None, 
                          pad_inches=0.2,
                          metadata=None)
```
